```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]  
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
time_step = 62
X_train, y_train = create_dataset(train_data, time_step)
```

```python
X_train.shape
```

```python
X_test, y_test = create_dataset(test_data, time_step)
```

```python
print(X_train.shape), print(y_train.shape), print(X_test.shape), print(y_test.shape)
```

```python
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM, ReLU
```

```python
X_train.shape
```

```python
lstm = Sequential()
lstm.add(LSTM(100, input_shape=(62,1), return_sequences=True))
lstm.add(LSTM(100, return_sequences=False))
lstm.add(Dense(1))
```

```python
lstm.compile(loss='mean_squared_error',optimizer='adam')
```

```python
lstm.summary()
```

```python
lstm.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,batch_size=64,verbose=1)
```
